# DBSCAN Clustering - Scikit-Learn
## Density-Based Spatial Clustering of Applications with Noise

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Extract the dataset values
X, label = datasets.make_moons(n_samples=200, noise=0.1)
print('The shape of the dataset: {0}'.format(X.shape))
```

```python
# Plot the dataset
plt.scatter(X[:,0],X[:,1])
plt.title('Dataset Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Load the model
"""
eps:
float, default=0.5
The maximum distance between two samples for one to be considered as in the neighborhood
of the other. This is not a maximum bound on the distances of points within a cluster.
This is the most important DBSCAN parameter to choose appropriately for your data set and distance function.

min_samples:
int, default=5
The number of samples (or total weight) in a neighborhood for a point to be considered as
a core point. This includes the point itself.
"""
model = DBSCAN(eps=0.25, min_samples=10)
model.fit(X)
```

```python
# Compare the model's labels against the dataset labels
plt.scatter(X[:,0],X[:,1],c=model.labels_)
plt.title('Model Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

plt.scatter(X[:,0],X[:,1],c=label)
plt.title('Dataset Label Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Extract the dataset values
X, label = datasets.make_blobs(n_samples=500, random_state=8)
print('The shape of the dataset: {0}'.format(X.shape))
```

```python
# Plot the dataset
plt.scatter(X[:,0],X[:,1])
plt.title('Dataset Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Load the model
model = DBSCAN(eps=1.5, min_samples=5)
model.fit(X)
```

```python
# Compare the model's labels against the dataset labels
plt.scatter(X[:,0],X[:,1],c=model.labels_)
plt.title('Model Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

plt.scatter(X[:,0],X[:,1],c=label)
plt.title('Dataset Label Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Define the dataset as a pandas dataframe
df = pd.DataFrame({
    'x': [15,20,25,30,28,12,20,28,18,29,33,24,45,45,52,51,52,55,53,55,48,51,55,45,50,61,64,69,72,55,60,65,70,68],
    'y': [35,40,45,50,43,39,36,30,52,54,46,55,59,63,70,66,63,58,60,65,70,68,62,23,14, 8,19, 7,24,20,15,10,12, 9]
})
xminmax = (min(df['x']),max(df['x']))
yminmax = (min(df['y']),max(df['y']))
print('X min: %d and X max: %d' %(xminmax[0],xminmax[1]))
print('Y min: %d and Y max: %d' %(yminmax[0],yminmax[1]))
```

```python
# Plot the dataset
plt.scatter(df['x'],df['y'])
plt.title('Dataset Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Load the model
model = DBSCAN(eps=10, min_samples=2) # eps uses the data's scale
model.fit(df)
```

```python
# Visualize the clustering output
plt.scatter(df['x'],df['y'],c=model.labels_)
plt.title('Model Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```
